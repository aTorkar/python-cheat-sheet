## while loop

Python while loop syntax:
```python
while expression:
    statement(s)
```

````python
count = 0
while count < 3:
    print(count)
    count += 1

# single statement while block
new_count = 0
while (new_count < 2): new_count += 1; print("Hard work")
````

````python
# checks if list is not empty, prints last element and removes it from the list
a_list = [7, 5, 654, 1]
while a_list:
    print(a_list.pop())
````

### switch case (without if-else blocks)

<div class="alert alert-block alert-info">
    <li>easier to debug and understand</li>
    <li>is a replacement of the ‘if..else’ statement in any program when we <b>know</b> that the choices are going to be <b>integer or character literals</b></li>
    <li>alternatives: if-else, dictionaries, mapping...</li>
</div>

````python
def case1():
    print("This is case 1")
    
def case2():
    print("This is case 2")
    
def default():
    print("This is the default case ")
    
def switch_case(case):
    switch = {
        1: case1,
        2: case2
    }
    return switch.get(case, default)()

# print(type(switch_case(1)))
switch_case(1)
switch_case(2)
switch_case(7)
````

### switch case with match keyword

Syntax for switch case with match keyword:
```python
match subject:
    case <pattern_1>:
        <action_1>
    case <pattern_2>:
        <action_2>
    case _:
        <action_wildcard>
```

````python
user_input = int(input("enter your lucky NUMBER between 1-3 (including)\n"))

match user_input:
    case 1:
        print("You will have a new house.")
    case 2:
        print("You will receive good news.")
    case 3:
        print("You will receive a kitten.")
    case _:
        print("You will receive a BOOK with reading lessons.")
````

## try/except

<div class="alert alert-block alert-info">
    <li><b>try</b>: used to enclose whatever might raise an exception</li>
    <li><b>except</b>: is used in conjunction with a try block. It allows you to define the exception type and provide appropriate error handling code</li>
    <li><b>finally</b>: is used in conjunction with a try-except block. Provides a mechanism to execute cleanup code, regardless of whether an exception occurred or not. It's always executed.</li>
    <li><b>raise</b>: is used to explicitly raise an exception in your code. It allows you to create custom exceptions or propagate built-in exceptions to higher levels of the program for handling.</li>
</div>

<div>
    
</div>

````python
# raise current exception 
try:
    a = int("67")
    print(a)
    x = int("abc")
    print(x)
except (ValueError, TypeError):
    print("Invalid conversion! Not a number!")
    
#########
try:
    d = 4/2
    print(d)
    c = 4/0
    print(c)
except Exception as e:
    print("Error:", str(e))
````

````python
try:
    k = 5//0 # raises divide by zero exception.
    print(k)
    
# examples: ZeroDivisionError, NameError, TypeError...
# handles zerodivision exception    
except ZeroDivisionError:
    print("Can't divide by zero")
        
finally:
    # this block is always executed 
    # regardless of exception generation.
    print('This is always executed') 
````

<b>raise: will re-raise error after the except</b>

```python
try:
    print(7/0)
except ZeroDivisionError as err:
    print("Logging excpetion:", str(err))
    raise
```

````python
try: 
    try:
        print(7/0)
    except ZeroDivisionError as err:
        print("Logging excpetion:", str(err))
        raise ValueError ("It is negative")
except Exception as err:
    print("Logging excpetion:", str(err))
````

````python
# handle the lists
# create lists
nums = [1, 2, 3]
words = ["kitty", "cow", "horse"]

# access 2nd element (1st element has index 0)
print(words[1])

# modify list
nums[0] = 77
print(nums)
````

````python
# slicing
numbers = [1, 2, 3, 4, 5]
print(numbers[1:4])
print(numbers[:3])
print(numbers[2:])
````

````python
# list methods 
fruits = ["apple", "banana", "orange"]
fruits.append("kiwi")
print(fruits)
fruits.insert(1, "grape")
print(fruits)
fruits.remove("banana")
print(fruits)
````

````python
# list comprehension
numbers = [1, 2, 3, 4, 5]
squares = [x ** 2 for x in numbers]
even_numbers = [x for x in numbers if x % 2 == 0] 
print(squares)
print(even_numbers)
print(len(even_numbers)) # print length (number of elements)
````

````python
# sorting
numbers = [5, 3, 1, 4, 2]
numbers2 = [5, 3, 1, 2]
numbers.sort()  # Sort the list in ascending order
print(numbers)
sorted_numbers = sorted(numbers2)  # Create a new sorted list
print(sorted_numbers)
````

````python
# reversing a list
numbers = [1, 2, 3, 4, 5]
numbers.reverse()
print(numbers)

numbers2 = [5, 6, 7, 8, 9]
reversed_numbers = numbers2[::-1]
print(reversed_numbers)
````
